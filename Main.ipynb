```python
def plot_comparison(n, NN, SVM, KNN, DT, BT, title, xlabel, ylabel):
    
    plt.figure()
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.plot(n, DT, '-', color="m", label="Decision Tree")
    plt.plot(n, NN, '-', color="b", label="Neural Network")
    plt.plot(n, KNN, '-', color="g", label="K-Nearest Neighbors")
    plt.plot(n, BT, '-', color="k", label="Boosted Decision Tree")
    plt.plot(n, SVM, '-', color="r", label="Support Vector Machines")
    plt.legend(loc="best", fontsize=7)
    plt.annotate('fjb7@gatech.edu', xy=(0.3, 0.3), xycoords='axes fraction', rotation=45, alpha=0.5, fontsize=25, color='gray')
    plt.show()

plot_comparison(
    train_samples_spam, 
    nn_fit_time_spam, 
    svm_fit_time_spam, 
    knn_fit_time_spam, 
    dt_fit_time_spam, 
    bt_fit_time_spam, 
    "Model Training Times: Spam Dataset",
    ylabel="Fit/Training Time (s)",
    xlabel="Number of Examples"
)              
plot_comparison(
    train_samples_spam, 
    nn_pred_time_spam, 
    svm_pred_time_spam, 
    knn_pred_time_spam,                  
    dt_pred_time_spam, 
    bt_pred_time_spam, 
    "Model Prediction Times: Spam Dataset",
    ylabel="Prediction Time (s)",
    xlabel="Number of Examples"
)   
plot_comparison(
    train_samples_spam, 
    nn_train_score_spam, 
    svm_train_score_spam, 
    knn_train_score_spam, 
    dt_train_score_spam, 
    bt_train_score_spam, 
    "Model Learning Rates: Spam Dataset",
    ylabel="Performance Score (F1)",
    xlabel="Number of Examples"
)  

plot_comparison(
    train_samples_cancer, 
    nn_fit_time_cancer, 
    svm_fit_time_cancer, 
    knn_fit_time_cancer, 
    dt_fit_time_cancer, 
    bt_fit_time_cancer, 
    "Model Training Times: Breast Cancer Dataset",
    ylabel="Fit/Training Time (s)",
    xlabel="Number of Examples"
)              
plot_comparison(
    train_samples_cancer, 
    nn_pred_time_cancer, 
    svm_pred_time_cancer, 
    knn_pred_time_cancer,                  
    dt_pred_time_cancer, 
    bt_pred_time_cancer, 
    "Model Prediction Times: Breast Cancer Dataset",
    ylabel="Prediction Time (s)",
    xlabel="Number of Examples"
)   
plot_comparison(
    train_samples_cancer, 
    nn_train_score_cancer, 
    svm_train_score_spam, 
    knn_train_score_cancer, 
    dt_train_score_cancer, 
    bt_train_score_cancer, 
    "Model Learning Rates: Breast Cancer Dataset",
    ylabel="Performance Score (F1)",
    xlabel="Number of Examples"
)
```
